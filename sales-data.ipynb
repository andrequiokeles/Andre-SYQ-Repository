```python
#importing the needed libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#opening the needed dataframes
csvdf = pd.read_csv('fct_invoice.csv')
jsondf = pd.read_json('dim_customer.json')
```

```python
'''
EASY 1
How many unique customers are in the dataset?
'''
uc = csvdf['customer_id'].nunique()
print(f"There are {uc} unique customers in the dataset.")
```

```python
'''
EASY 2
What are the different categories of products available? How many unique categories are there?
'''
dc = csvdf['category'].unique()
uc = csvdf['category'].nunique()
print(f"The different categories of products available are: {', '.join(dc)}.")
print(f"There are {uc} unique categories.")
```

```python
'''
EASY 3
Which payment method is the most popular? How many times was it used?
'''
payment_list = csvdf['payment_method'].value_counts()
mpp = payment_list.head(1)
print(f"The most popular payment method is {str(mpp.index[0])}, and it was used {int(mpp.values[0])} times.")
```

```python
'''
MEDIUM 1
What are the three most popular categories, by total sales?
'''
csvdf['total_sales'] = csvdf['quantity']*csvdf['price']
mpc = csvdf.groupby('category')['total_sales'].sum()
top_categories = mpc.sort_values(ascending=False).head(3)
print(f"The top 3 categories are: {', '.join(top_categories.index)}. ")
print(top_categories)
```

```python
'''
MEDIUM 2
What are the total sales attributed to customers over the age of 45?
'''
med2table = pd.merge(csvdf, jsondf, left_on = 'customer_id', right_on = 'id', how = 'outer')
over45people = med2table[med2table['age']>45]
tos_over45 = over45people['total_sales'].sum()
print(f"The total sales attributed to customers over the age of 45 is: {tos_over45}.")
```

```python
'''
MEDIUM 3
How is the data distributed across different invoice dates? 
Are there any seasonal trends or patterns? (Use a graph for this.)
'''
csvdf['invoice_date'] = pd.to_datetime(csvdf['invoice_date'], format = '%d/%m/%Y')
graph_basis = csvdf.groupby('invoice_date')['total_sales'].sum()

plt.figure(figsize=(16,10))
plt.plot(graph_basis.index, graph_basis.values, marker='o', linestyle='-', color='b', label='Sales')  
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Sales Over Time')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print('There are no seasonal trends or patterns.')
```

```python
'''
HARD 1
Create a pivot table showing the breakdown of sales across these dimensions, 
in this order: category, decade age range (e.g., 10-19, 20-29, and so forth)
'''
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pt_data = (pd.merge(csvdf, jsondf, left_on = 'customer_id', right_on = 'id', how = 'outer')).sort_values('age').reset_index()
pt_data['decade_age_range'] = pd.cut(pt_data['age'], bins=range(0, 101, 10), labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99'])
final_pivot = pd.pivot_table(pt_data, values = 'total_sales', index = ['category','decade_age_range'], aggfunc = ['sum', 'mean'], observed = False)

final_pivot
```
